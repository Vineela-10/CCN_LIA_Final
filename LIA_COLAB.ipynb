```python
from google.colab import files
import os
from IPython.display import Video
import glob
import shutil

print("üì∏ Upload your SOURCE IMAGE (.jpg, .png)")
print("This is the face/person you want to animate")
uploaded = files.upload()
source_file = list(uploaded.keys())[0]

source_clean = "source_image.jpg"
shutil.copy(source_file, source_clean)
print(f"‚úÖ Source ready\n")

print("üé¨ Upload your DRIVING VIDEO (.mp4, .avi)")
print("This provides the motion/animation")
uploaded = files.upload()
driving_file = list(uploaded.keys())[0]

driving_clean = "driving_video.mp4"
shutil.copy(driving_file, driving_clean)
print(f"‚úÖ Driving video ready\n")

print("üöÄ Processing... (this may take 2-5 minutes)")
os.system(f'python run_demo.py --model vox --source_path {source_clean} --driving_path {driving_clean}')

results = glob.glob('/content/LIA/res/vox/*.mp4')
if results:
    latest = max(results, key=os.path.getctime)

    print("\n‚úÖ Animation complete!")
    display(Video(latest, width=600, embed=True))

    print("\nüíæ Downloading your result...")
    files.download(latest)
else:
    print("‚ùå No result found. Check errors above.")
```
